```python
from datetime import datetime

import pandas as pd
```

```python
noahs_customers = pd.read_csv("../noahs-customers.csv")

noahs_orders_items = pd.read_csv("../noahs-orders_items.csv")

noahs_orders = pd.read_csv("../noahs-orders.csv")

noahs_products = pd.read_csv("../noahs-products.csv")
```

```python
# Format the orders DataFrame

noahs_orders['shipped'] = pd.to_datetime(noahs_orders['shipped'])

noahs_orders["shipping_date"] = noahs_orders["shipped"].apply(
    lambda x: x.strftime("%Y-%m-%d"))
```

```python
noahs_customers_and_orders = pd.merge(noahs_customers, noahs_orders, how="inner", on="customerid")

noahs_products_and_orders_items = pd.merge(noahs_products, noahs_orders_items, how="inner", on="sku")

noahs_aggregated_df = pd.merge(noahs_customers_and_orders, noahs_products_and_orders_items, how="inner", on="orderid")
```

```python
# Find the customer id of the person from previous puzzle

initial_person_customerid = noahs_customers[noahs_customers["phone"]
                                            == "585-838-9161"]["customerid"].iloc[0]

initial_person_customerid
```

```python
noahs_aggregated_df_initial_person_customerid = noahs_aggregated_df[noahs_aggregated_df["customerid"] == initial_person_customerid]

noahs_aggregated_df_initial_person_customerid
```

```python
colours = [
    "red", "blue", "green", "yellow", "purple",
    "orange", "pink", "brown", "black", "white",
    "gray", "cyan", "magenta", "teal", "indigo",
    "violet", "gold", "silver", "beige", "maroon"
]


def check_desc_colours(desc_name: str) -> str | None:

    for colour in colours:

        if colour in desc_name:

            return colour

    return None
```

```python
noahs_aggregated_df_initial_person_customerid["colour"] = noahs_aggregated_df_initial_person_customerid["desc"].apply(check_desc_colours)

noahs_aggregated_df_initial_person_customerid_colour = noahs_aggregated_df_initial_person_customerid.dropna(subset=["colour"])

noahs_aggregated_df_initial_person_customerid_colour
```

```python
def extract_item_desc(desc_name: str) -> str:

    if "(" not in desc_name or ")" not in desc_name:

        return extract_item_desc
    
    open_bracket_index = desc_name.index("(")

    return desc_name[open_bracket_index:]
```
