```python
import pandas as pd
```

```python
customer_data = pd.read_csv("../noahs-customers.csv")

orders_items_data = pd.read_csv("../noahs-orders_items.csv")

orders_data = pd.read_csv("../noahs-orders.csv")

products_data = pd.read_csv("../noahs-products.csv")
```

```python
customer_data = customer_data[["customerid", "name"]]

orders_data = orders_data[["orderid", "customerid", "ordered", "shipped"]]
```

```python
customer_data.head()
```

```python
orders_data.head()
```

```python
customer_data_name_split = customer_data["name"].str.split()

customer_data["first_name"] = customer_data_name_split.apply(lambda a: a[0])

customer_data["last_name"] = customer_data_name_split.apply(lambda a: a[1])
```

```python
customer_data = customer_data[["customerid", "first_name", "last_name"]]

customer_data = customer_data[customer_data["first_name"].str.startswith("J") & customer_data["last_name"].str.startswith("P")]

customer_data
```

```python
customer_order_df = pd.merge(customer_data, orders_data, how="inner", on="customerid")
```

```python
customer_order_df.head()
```

```python
orders_items_data = orders_items_data[["orderid", "sku"]]

products_data = products_data[["sku", "desc"]]
```

```python
customer_order_items_df = pd.merge(customer_order_df, orders_items_data, how="inner", on="orderid")
```

```python
customer_order_items_desc_df = pd.merge(customer_order_items_df, products_data, how="inner", on="sku")
```

```python
customer_order_items_desc_df.head()
```

```python
customer_order_agg_df = customer_order_items_desc_df[["customerid", "first_name", "last_name", "ordered", "shipped", "desc"]]
```

```python
customer_order_agg_df.head()
```

```python
customer_order_agg_df = customer_order_agg_df[customer_order_agg_df["ordered"].str.startswith("2017") | customer_order_agg_df["shipped"].str.startswith("2017")]
```
