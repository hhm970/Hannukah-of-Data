```python
import pandas as pd
```

```python
customer_data = pd.read_csv("../noahs-customers.csv")

orders_items_data = pd.read_csv("../noahs-orders_items.csv")

orders_data = pd.read_csv("../noahs-orders.csv")

products_data = pd.read_csv("../noahs-products.csv")
```

```python
customer_data = customer_data[['customerid', 'name', 'phone']]

orders_data = orders_data[['orderid', 'customerid', 'ordered']]

orders_items_data = orders_items_data[['orderid', 'qty', 'sku']]

products_data = products_data[['sku', 'desc']]
```

```python
# Filter orders data

orders_data["ordered"] = pd.to_datetime(orders_data["ordered"])

start_time = "00:00:00"

end_time = "05:00:00"

valid_days = {0, 1, 2, 3, 6}


def day_in_list(input_day: str) -> bool:
    if input_day in valid_days:
        return True
    return False

orders_data = orders_data[(orders_data["ordered"].dt.time >= pd.Timestamp(start_time).time()) & (orders_data["ordered"].dt.time <= pd.Timestamp(end_time).time())]

orders_data = orders_data[orders_data["ordered"].dt.day_of_week.apply(day_in_list) == True]

orders_data.head()
```

```python
# Filter orders_items data

orders_items_data = orders_items_data[orders_items_data["qty"] > 1]
```

```python
# Filter products_data

products_data = products_data[products_data['sku'].str.contains("BKY")]
```

```python
customer_orders_data = pd.merge(customer_data, orders_data, how='inner', on='customerid')

product_orders_items_data = pd.merge(products_data, orders_items_data, how='inner', on='sku')

aggregate_data = pd.merge(customer_orders_data, product_orders_items_data, how='inner', on='orderid')
```
