```python
import pandas as pd
```

```python
noahs_customers = pd.read_csv("../noahs-customers.csv")

noahs_orders_items = pd.read_csv("../noahs-orders_items.csv")

noahs_orders = pd.read_csv("../noahs-orders.csv")

noahs_products = pd.read_csv("../noahs-products.csv")
```

```python
noahs_customers = noahs_customers[["customerid", "phone"]]

noahs_orders = noahs_orders[["orderid", "customerid"]]

noahs_products = noahs_products[["sku", "desc", "wholesale_cost"]]
```

```python
noahs_customers_orders = pd.merge(noahs_customers, noahs_orders, how="inner", on="customerid")
```

```python
noahs_products_orders_items = pd.merge(noahs_products, noahs_orders_items, how="inner", on="sku")
```

```python
noahs_df = pd.merge(noahs_customers_orders, noahs_products_orders_items, how="inner", on="orderid")

noahs_df.head()
```

```python
noahs_df["profit"] = noahs_df["qty"] * (noahs_df["unit_price"] - noahs_df["wholesale_cost"])
```

```python
noahs_df = noahs_df[["phone", "profit"]]
```

```python
noahs_df.sort_values(by="profit", ascending=True)
```
