```python
from datetime import datetime

import pandas as pd
```

```python
noahs_customers = pd.read_csv("../noahs-customers.csv")

noahs_orders_items = pd.read_csv("../noahs-orders_items.csv")

noahs_orders = pd.read_csv("../noahs-orders.csv")

noahs_products = pd.read_csv("../noahs-products.csv")
```

```python
noahs_customers = noahs_customers[['customerid', 'name', 'phone']]

noahs_orders = noahs_orders[['orderid', 'customerid', 'shipped']]

noahs_orders_items = noahs_orders_items[['orderid', 'sku', 'qty']]

noahs_products = noahs_products[['sku', 'desc']]
```

```python
noahs_orders.dtypes
```

```python
# Find the customer id of the person from previous puzzle

initial_person_customerid = noahs_customers[noahs_customers["phone"] 
                                            == "585-838-9161"]["customerid"].iloc[0]

initial_person_customerid
```

```python
noahs_orders['shipped'] = pd.to_datetime(noahs_orders['shipped'])
```

```python
# Find the date from shipped column

noahs_orders["shipping_date"] = noahs_orders["shipped"].apply(lambda x: x.strftime("%Y-%m-%d"))
```

```python
noahs_orders
```

```python
noahs_ordered_products = noahs_products.merge(noahs_orders_items, on="sku", how="inner")

noahs_ordered_products
```

```python
noahs_ordered_products_customer_id = noahs_ordered_products.merge(noahs_orders, on="orderid", how="inner")

noahs_ordered_products_customer_id
```

```python
noahs_ordered_products_customer_id["duplicated_desc_date"] = noahs_ordered_products_customer_id.duplicated(
    subset=["desc", "shipping_date"], keep=False)

noahs_ordered_products_customer_id_duplicates = noahs_ordered_products_customer_id[noahs_ordered_products_customer_id["duplicated_desc_date"] == True]
```

```python
noahs_ordered_products_customer_id_duplicates.head()
```

```python
colours = [
    "red", "blue", "green", "yellow", "purple",
    "orange", "pink", "brown", "black", "white",
    "gray", "cyan", "magenta", "teal", "indigo",
    "violet", "gold", "silver", "beige", "maroon"
]

def check_desc_colours(desc_name: str) -> str | None:

    for colour in colours:

        if colour in desc_name:

            return colour
        
    return None
```

```python
noahs_ordered_products_customer_id_duplicates["product_coloured"] = noahs_ordered_products_customer_id_duplicates["desc"].apply(check_desc_colours)
```

```python
noahs_ordered_products_customer_id_duplicates.head()
```

```python
noahs_ordered_products_customer_id_duplicates.dropna(subset=["product_coloured"], inplace=True)

noahs_ordered_products_customer_id_duplicates["product_coloured"].value_counts()
```

```python
noahs_ordered_products_customer_details_duplicates = noahs_ordered_products_customer_id_duplicates.merge(noahs_customers, on="customerid", how="inner")

noahs_ordered_products_customer_details_duplicates[
    noahs_ordered_products_customer_details_duplicates["customerid"] == initial_person_customerid]
```

```python
unique_dates = list(noahs_ordered_products_customer_details_duplicates["shipping_date"].unique())

unique_dates
```

```python
result = []

for date in unique_dates:

    iter_df = noahs_ordered_products_customer_details_duplicates[
        noahs_ordered_products_customer_details_duplicates["shipping_date"] == date]

    if initial_person_customerid in list(iter_df["customerid"].unique()):

        result.append(date)

result
```
